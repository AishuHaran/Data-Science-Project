### **Question 3 - How many passengers of each sex survived and how many of them did not? What is the survival rate for passengers of each sex?**



```python
# Write Your Code here
a = sns.catplot(
    x="sex", y="survived", hue="class", kind="bar",
    hue_order=["First", "Second", "Third"], ci=None, palette="rocket", data=df)
a.fig.suptitle("Titanic Survivors", fontsize=12)
print()
overall_survival_rate = df['survived'].mean() * 100
print(f"Overall Survival Rate: {overall_survival_rate:.2f}%")
```

### **Question 4 - For how many passengers do we have age information (including estimated age)? For how many is the age information missing? What is the age distribution for passengers whose age information is available?**


```python
#Write Your Code here
#For how many passengers do we have age information (including estimated age)?
df
```

```python
age_count = df['age'].notnull().sum()
print(f"Number of passengers with age information: {age_count}")
```

```python
#For how many is the age information missing?
age_missing = df['age'].isnull().sum()
print(f"Age information is missing for {age_missing} passengers")
```

```python
#What is the age distribution for passengers whose age information is available?
age_count = df[df['age'].notnull()]
age_count
```

```python
#Graph for age distribution for passengers whose age information is available
sns.histplot(age_count['age'], bins=10, kde=True, color="pink")
plt.title('Age Distribution of Titanic Passengers', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Passengers', fontsize=14)
plt.grid(True)
```

```python
# Age, subset sex
palette = sns.color_palette("Paired")
selected_colors = [palette[2], palette[9]]
sns.histplot(
    data=age_count,          # Data source
    x='age',                 # Variable for the x-axis
    hue='sex',               # Variable to color by sex
    bins=10,                 # Number of bins
    kde=True,                # Include Kernel Density Estimate
    palette=selected_colors  # Use the selected colors for the hue
)
plt.title('Age Distribution of Titanic Passengers by Sex', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Passengers', fontsize=14)
plt.grid(True)
```
